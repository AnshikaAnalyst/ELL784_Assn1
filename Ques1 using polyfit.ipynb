```python
import pandas as pd
import numpy as np
train_data=pd.read_csv(r'C:\Users\garga\3D Objects\ELL784 Assn\Assn1\train.csv')

train_data.head()
```

```python
test_data=pd.read_csv(r'C:\Users\garga\3D Objects\ELL784 Assn\Assn1\test.csv',header=None)

test_data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
train_data=scaler.fit_transform(train_data)
train_data=pd.DataFrame(train_data)
```

```python
x_train=train_data.iloc[:,0]
y_train=train_data.iloc[:,1]
```

```python
x_test=test_data.iloc[:,0]
y_test=test_data.iloc[:,1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train)
```

```python
#taking only first 20 points of x and fit polynomial curve by using polyfit function
from sklearn.metrics import mean_squared_error, r2_score
M=4
x_20=x_train[0:20]
y_20=y_train[0:20]
x_20_sorted=np.sort(x_20)
f1=np.polyfit(x_20,y_20,1)
f2=np.polyfit(x_20,y_20,2)
f3=np.polyfit(x_20,y_20,3)
f4=np.polyfit(x_20,y_20,4)

for coefs in [f1,f2,f3,f4]:
    poly=np.poly1d(coefs)
    print('Coefficients:',coefs[:-1])
    print('Intercept:',coefs[-1])
    print('Error:',r2_score(y_20,np.poly1d(coefs)(x_20)))
    

    fitted_curve = np.poly1d(coefs)(x_20_sorted)
    #y_pred=np.poly1d(coefs)(x_20)
    #fitted_curve_sorted=np.sort(fitted_curve)
    plt.scatter(x_20, y_20, label="observed")
    plt.plot(x_20_sorted, fitted_curve, c="red", label="fitted")
    plt.grid()
    plt.legend()
    plt.show()
```

```python
#taking full training data set and fit polynomial curve by using polyfit function
from sklearn.metrics import mean_squared_error, r2_score
x_train_sorted=np.sort(x_train)
f1=np.polyfit(x_train,y_train,1)
f2=np.polyfit(x_train,y_train,2)
f3=np.polyfit(x_train,y_train,3)
f4=np.polyfit(x_train,y_train,4)

for coefs in [f1,f2,f3,f4]:
    poly=np.poly1d(coefs)
    print('Coefficients:',coefs[:-1])
    print('Intercept:',coefs[-1])
    print('Error:',r2_score(y_train,np.poly1d(coefs)(x_train)))
    

    fitted_curve = np.poly1d(coefs)(x_train_sorted)
    #fitted_curve_sorted=np.sort(fitted_curve)
    plt.scatter(x_train, y_train, label="observed")
    plt.plot(x_train_sorted, fitted_curve, c="red", label="fitted")
    plt.grid()
    plt.legend()
    plt.show()
```

```python
# for higher degree of polynomials
from sklearn.metrics import mean_squared_error, r2_score
x_train_sorted=np.sort(x_train)
f1=np.polyfit(x_train,y_train,5)
f2=np.polyfit(x_train,y_train,6)
f3=np.polyfit(x_train,y_train,7)
f4=np.polyfit(x_train,y_train,8)

for coefs in [f1,f2,f3,f4]:
    poly=np.poly1d(coefs)
    print('Coefficients:',coefs[:-1])
    print('Intercept:',coefs[-1])
    print('Error:',r2_score(y_train,np.poly1d(coefs)(x_train)))
    

    fitted_curve = np.poly1d(coefs)(x_train_sorted)
    #fitted_curve_sorted=np.sort(fitted_curve)
    plt.scatter(x_train, y_train, label="observed")
    plt.plot(x_train_sorted, fitted_curve, c="red", label="fitted")
    plt.grid()
    plt.legend()
    plt.show()
```

### we can clearly see that for higher degree of polynomial it is overfitting

# test the model for test data

```python
#taking only first 20 points of x and fit polynomial curve by using polyfit function
from sklearn.metrics import mean_squared_error, r2_score
M=3
x_20=x_train[0:20]
y_20=y_train[0:20]
b=np.polyfit(x_20,y_20,M)
poly=np.poly1d(b)
#s=np.sort(x_20)
#y_pred=poly(s)
plt.plot(np.sort(x_20),poly(np.sort(x_20)),x_20,y_20,'o')
plt.legend(('model','training data'))
# compute rmse
#error=((1/len(y_pred))*(np.sum((y_pred-y_20)**2)))**0.5
print(r2_score(y_20,poly(x_20)))
```

```python
# try fitting whole training data
M=3
poly=np.poly1d(np.polyfit(x_train,y_train,M))
plt.plot(np.sort(x_train),poly(np.sort(x_train)),x_train,y_train,'o')
plt.legend(('model','complete training data'))
# compute rmse
print(r2_score(y_train,poly(x_train)))
```
